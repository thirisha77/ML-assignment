```python
import numpy as np  
import matplotlib.pyplot as mtp  
import pandas as pd  
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error, r2_score

# Load the student performance dataset
file_path = 'Student_Performance.csv'
student_data = pd.read_csv(file_path)

# Define the independent variables (Hours Studied, Previous Scores, Sleep Hours, Sample Question Papers Practiced) 
# and dependent variable (Performance Index)
x = student_data[['Hours Studied', 'Previous Scores', 'Sleep Hours', 'Sample Question Papers Practiced']].values
y = student_data['Performance Index'].values

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

# Create a linear regression model and train it
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Predict performance for the test and train sets
y_pred = regressor.predict(x_test)
x_pred = regressor.predict(x_train)

# Plot the training set results (using Hours Studied for visualization)
mtp.scatter(x_train[:, 0], y_train, color="green")
mtp.plot(x_train[:, 0], x_pred, color="red")
mtp.title("Linear Regression on Student Performance Dataset (Training Set)")
mtp.xlabel("Hours Studied")
mtp.ylabel("Performance Index")
mtp.show()

# Plot the testing set results
mtp.scatter(x_test[:, 0], y_test, color="blue")
mtp.plot(x_train[:, 0], x_pred, color="red")
mtp.title("Linear Regression on Student Performance Dataset (Testing Set)")
mtp.xlabel("Hours Studied")
mtp.ylabel("Performance Index")
mtp.show()

# Output model coefficients and performance
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Coefficient of Determination (R^2 Score):', r2_score(y_test, y_pred))
```
